```python

def print_category_counts(df, col):
    print(col, "counts:")
    for c in df[col].unique():
        count = (df[col] == c).sum()
        part = count / len(df)
        print(f"{c:>10s}: {count} ({part:2.1%})")
```

```python
df = pd.read_csv(root_dir + "/data/diabetes-130-us/diabetic_data.csv", na_values="?", dtype={'payer_code': str})
print("DataFarame size", df.shape)
print_category_counts(df, "readmitted")
df.head(5)
```

```python
print("Missing values (%):")
for col in df.columns:
    percent_missing = df[col].isna().mean() * 100
    if percent_missing > 0:
        print(col, " : ", percent_missing)
```
